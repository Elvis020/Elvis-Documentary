# Broadcasting

```python
import numpy as np
import csv
np.set_printoptions(formatter={'float': '{: 6.1f}'.format})
```

```python
with open('AAPL-2017.csv', newline='') as file2017:
    reader1 = csv.DictReader(file2017)
    array2017_open_and_close = np.empty((0,2))
    for row in reader1:
        oneRow17 = np.array([[row['Open'],row['Close']]])
        array2017_open_and_close = np.append(array2017_open_and_close,oneRow17,axis=0)
print(array2017_open_and_close)
```

```python
# Projecting Open and Close column into the future
projection = np.array(array2017_open_and_close,dtype=np.float32)
projection_open = projection[:,0] * 1.1
projection_close = projection[:,1] * 1.2
```

```python
# Before broadcasting
print(projection[:,0][:5])  #open
# print(projection[:,1][:5])  #close
```

```python
# After broadcasting
print(projection_open[:5])
# print(projection_close[:5])
```

# Numpy Array Indexing and Slicing

```python
import numpy as np
import csv
```

```python
with open('AAPL-2016.csv', newline='') as file2016:
    reader1 = csv.DictReader(file2016)
    array2016 = np.empty((0,2))
    for row in reader1:
        oneRow16 = np.array([[row['Open'],row['Volume']]])
        array2016 = np.append(array2016,oneRow16,axis=0)
print(array2016)
```

```python
with open('AAPL-2017.csv',newline='') as file2017:
    reader2 = csv.DictReader(file2017)
    array2017 = np.empty((0,2))
    for row in reader2:
        oneRow17 = np.array([[row['Open'],row['Volume']]])
        array2017 = np.append(array2017,oneRow17,axis=0)
print(array2017)
```

```python
with open('AAPL-2016.csv', newline='') as file2016:
    reader1 = csv.DictReader(file2016)
    array2016_high_and_low = np.empty((0,2))
    for row in reader1:
        oneRow16 = np.array([[row['High'],row['Low']]])
        array2016_high_and_low = np.append(array2016_high_and_low,oneRow16,axis=0)
print(array2016_high_and_low)
```

```python
vertical_slice = np.array((array2016_high_and_low))


vertical_slice_of_high = vertical_slice[:,0][:5]
vertical_slice_of_low = vertical_slice[:,1][:5]
print("Individually")
print(vertical_slice_of_high)
print(vertical_slice_of_low)


print("Together") #The correct one
print(vertical_slice[:][:5])
```

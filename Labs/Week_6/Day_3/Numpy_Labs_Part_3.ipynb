## Operations on Arrays

```python
import numpy as np
import csv
```

## Question 1

```python
with open('AAPL-2016.csv', newline='') as file2016:
    reader1 = csv.DictReader(file2016)
    array2016 = np.empty((0,2))
    for row in reader1:
        oneRow16 = np.array([[row['Open'],row['Volume']]])
        array2016 = np.append(array2016,oneRow16,axis=0)
print(array2016)
```

```python
with open('AAPL-2017.csv',newline='') as file2017:
    reader2 = csv.DictReader(file2017)
    array2017 = np.empty((0,2))
    for row in reader2:
        oneRow17 = np.array(`[[row['Open'],row['Volume']]])
        array2017 = np.append(array2017,oneRow17,axis=0)
print(array2017)
```

---

```python
# Summing the volume and open for 2016
# Converting to float
sum_for_2016 = np.array(array2016).astype(np.float)

volume_and_new_stock_2016 = sum_for_2016.sum(axis=0)
print(volume_and_new_stock_2016)
```

```python
# Summing the volume and open for  2017
# Converting to float
sum_for_2017 = np.array(array2017).astype(np.float)
volume_and_new_stock_2017 = sum_for_2017.sum(axis=0)
print(volume_and_new_stock_2017)
```

```python
# Total sum for volume for both years
sum_of_volume_for_both_years = volume_and_new_stock_2016[1] + volume_and_new_stock_2017[1]
print(sum_of_volume_for_both_years)
```

```python
# Change in trade volume
volume_traded_in_2016 = volume_and_new_stock_2016[1]
volume_traded_in_2017 = volume_and_new_stock_2017[1]
change_in_trade_volume =  abs(volume_traded_in_2017 - volume_traded_in_2016)
change_in_trade_volume =  volume_traded_in_2017 - volume_traded_in_2016
print(change_in_trade_volume)
```

# Pandas Series

```python
import numpy as np
import pandas as pd
import csv
import datetime
import matplotlib.pyplot as plt
```

```python
with open('AAPL-2017.csv',newline='') as file2017:
    reader2 = csv.DictReader(file2017)
    array2017 = np.empty((0,2))
    for row in reader2:
        oneRow17 = np.array([[row['Date'],row['Volume']]])
        array2017 = np.append(array2017,oneRow17,axis=0)
print(array2017)
```

```python
date = [datetime.datetime.strptime(x,'%Y-%m-%d') for x in array2017[:,0]]
data = pd.Series(np.float64(array2017[:,1]), index=date)
data.plot()
plt.show()
```

# Working with Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
reading_data = pd.read_csv('AAPL-2017.csv', usecols=['Date','Volume'])[:40]
reading_data.plot()
plt.show()
```

```python
reading_open_of_2016 = pd.read_csv('AAPL-2016.csv', usecols=['Open'])
reading_open_of_2017 = pd.read_csv('AAPL-2017.csv', usecols=['Open'])
plt.plot(reading_open_of_2016,'b--',label='2016',color='red')
plt.plot(reading_open_of_2017,'b--',label='2017',color='blue')
plt.title('AAPL 2017 VS 2016')
plt.legend()
```

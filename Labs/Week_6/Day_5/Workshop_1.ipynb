## Inserting the csv file into a dataframe

```python
table_columns = ['first_name','last_name','address','city','county','state','zip','phone1','phone2']
df_for_500users = pd.DataFrame(df_data,columns=table_columns)
df_for_500users
```

```python
# Inserting into sql table
for row in df_for_500users.itertuples():
    cur.execute("insert into USERS500.users (first_name,last_name,address,city,country,state,zip) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s);",(row.first_name, row.last_name,row.address,row.city, row.county,row.state,row.zip, row.phone1,row.phone2))
con.commit()
```

## Question 6a

```python
# Selecting the number of people from each state
cur.execute("select state,count(state) from USERS500.users group by state order by count(state) desc")
rows = cur.fetchall()
for row in rows:
   print(row)
```

## Question 6b

```python
# Selecting the number of people from each state
cur.execute("select array_to_string(array_agg(first_name),' , ')  from USERS500.users group by zip")
rows = cur.fetchall()
for row in rows:
   print(row)
```

## Question 6c

```python
# Selecting the number of people from each state
cur.execute("select last_name::char as nummy,count(last_name::char) as nummy2 from USERS500.users u group by nummy order by nummy")
rows = cur.fetchall()
for row in rows:
   print(row)
```
